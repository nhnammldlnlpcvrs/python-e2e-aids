## 3. Làm Sạch Dữ Liệu (Data Cleaning)

Dữ liệu thực tế LUÔN LUÔN "bẩn". Làm sạch chiếm ~70% thời gian của Data Scientist.
Dưới đây là quy trình cơ bản.

### 3.1. Phát hiện Missing Values

```python
# ── isnull(): Trả về DataFrame True/False (True = bị thiếu) ─────────
print("=== MA TRẬN THIẾU (5 dòng đầu) ===")
display(df.isnull().head(10))  # Hiện 10 dòng đầu

# ── isnull().sum(): Đếm số giá trị thiếu TRONG TỪNG CỘT ─────────────
print("\n=== SỐ GIÁ TRỊ THIẾU MỖI CỘT ===")
missing = df.isnull().sum()
print(missing[missing > 0])  # Chỉ hiện cột có missing value

# ── isnull().sum().sum(): Tổng tất cả ô bị thiếu ────────────────────
print(f"\nTổng số ô bị thiếu trong toàn bộ bảng: {df.isnull().sum().sum()}")
```

### 3.2. Xử lý Missing Values

Có 2 chiến lược chính:
- **Xóa** (`dropna`): Khi missing value ít và không quan trọng
- **Điền** (`fillna`): Khi bạn có thể ước lượng giá trị thay thế

```python
# ── Tạo bản sao để không làm hỏng dữ liệu gốc ───────────────────────
df_clean = df.copy()

# ── Chiến lược 1: Điền giá trị thiếu (fillna) ───────────────────────
#   • Cột số (So_Luong, Gia_Ban): điền bằng MEDIAN (trung vị)
#     Dùng median thay vì mean vì median không bị ảnh hưởng bởi outlier

df_clean['So_Luong'] = df_clean['So_Luong'].fillna(df_clean['So_Luong'].median())
df_clean['Gia_Ban'] = df_clean['Gia_Ban'].fillna(df_clean['Gia_Ban'].median())

#   • Cột text (San_Pham, Danh_Muc): điền bằng "Không rõ"
df_clean['San_Pham'] = df_clean['San_Pham'].fillna('Không rõ')
df_clean['Danh_Muc'] = df_clean['Danh_Muc'].fillna('Không rõ')

#   • Cột boolean (Is_VIP): điền bằng False (coi như không phải VIP)
df_clean['Is_VIP'] = df_clean['Is_VIP'].fillna(False)

# ── Kiểm tra lại ────────────────────────────────────────────────────
print(f"Số ô thiếu sau khi xử lý: {df_clean.isnull().sum().sum()}")
print("✅ Tất cả missing value đã được xử lý!")

# ── Chiến lược 2: Xóa dòng chứa missing value (dropna) — CHỈ ĐỂ THAM KHẢO
# df_dropped = df.dropna()  # Xóa MỌI dòng có ít nhất 1 ô NaN
# print(f"Số dòng sau khi dropna: {len(df_dropped)}")
```

### 3.3. Phát hiện & Xóa Dòng Trùng Lặp (Duplicates)

```python
# ── duplicated(): Trả về Series True/False (True = dòng này trùng với dòng trước đó)
so_dong_trung = df_clean.duplicated().sum()
print(f"Số dòng bị trùng: {so_dong_trung}")

# ── Xem các dòng bị trùng ───────────────────────────────────────────
print("\n=== CÁC DÒNG BỊ TRÙNG ===")
display(df_clean[df_clean.duplicated(keep=False)])  # keep=False → hiện tất cả bản sao

# ── drop_duplicates(): Xóa dòng trùng, giữ lại bản đầu tiên ────────
df_clean = df_clean.drop_duplicates()
print(f"\nSố dòng sau khi xóa trùng: {len(df_clean)}")
print("✅ Dữ liệu đã sạch — không còn NaN, không còn trùng lặp!")
```

## 4. Lọc Dữ Liệu (Filtering)

Đây là kỹ năng sử dụng HÀNG NGÀY. Pandas cung cấp nhiều cách lọc: Boolean indexing, `loc`, `iloc`, `query()`.

### 4.1. Boolean Indexing — Cách phổ biến nhất

```python
# ── Dùng df_clean (đã làm sạch) cho các phần sau ────────────────────

# ── Lọc sản phẩm có giá bán > 500,000 ───────────────────────────────
gia_cao = df_clean[df_clean['Gia_Ban'] > 500_000]
print(f"=== SẢN PHẨM GIÁ > 500K ({len(gia_cao)} dòng) ===")
display(gia_cao.head(10))

# ── Lọc khách hàng VIP ──────────────────────────────────────────────
khach_vip = df_clean[df_clean['Is_VIP'] == True]
print(f"\n=== KHÁCH HÀNG VIP ({len(khach_vip)} dòng) ===")
display(khach_vip.head(10))

# ── Kết hợp nhiều điều kiện: & (AND), | (OR), ~ (NOT) ───────────────
#   ⚠️ PHẢI dùng dấu ngoặc đơn () cho từng điều kiện!
loc_dac_biet = df_clean[(df_clean['Gia_Ban'] > 500_000) & (df_clean['Is_VIP'] == True)]
print(f"\n=== KHÁCH VIP MUA HÀNG > 500K ({len(loc_dac_biet)} dòng) ===")
display(loc_dac_biet)

# ── Lọc theo danh mục: dùng isin() ──────────────────────────────────
dien_tu_thoi_trang = df_clean[df_clean['Danh_Muc'].isin(['Điện tử', 'Thời trang'])]
print(f"\n=== ĐIỆN TỬ + THỜI TRANG ({len(dien_tu_thoi_trang)} dòng) ===")
display(dien_tu_thoi_trang.head(10))
```

### 4.2. `loc` và `iloc` — Truy cập theo nhãn và vị trí

| Phương thức | Truy cập theo | Ví dụ |
|-------------|---------------|-------|
| `loc` | **Nhãn dòng/cột** (label) | `df.loc[0:5, 'San_Pham':'Gia_Ban']` |
| `iloc` | **Vị trí số nguyên** (integer position) | `df.iloc[0:5, 1:4]` |

```python
# ── loc[row_label, column_label] ────────────────────────────────────
print("=== df.loc[3:7, ['San_Pham', 'Gia_Ban', 'Is_VIP']] ===")
display(df_clean.loc[3:7, ['San_Pham', 'Gia_Ban', 'Is_VIP']])
# ⚠️ Với loc, khoảng [3:7] BAO GỒM CẢ dòng có index=7

# ── iloc[row_position, column_position] ────────────────────────────
print("\n=== df.iloc[3:8, 1:4] ===")
display(df_clean.iloc[3:8, 1:4])
# ⚠️ Với iloc, khoảng [3:8] KHÔNG BAO GỒM vị trí 8 (giống Python list)

# ── Lọc dòng bằng điều kiện + chọn cột bằng loc ─────────────────────
print("\n=== Tên + Giá của sản phẩm Điện tử ===")
display(df_clean.loc[df_clean['Danh_Muc'] == 'Điện tử', ['San_Pham', 'Gia_Ban']].head(10))
```

### 4.3. `query()` — Lọc bằng chuỗi điều kiện (dễ đọc hơn)

```python
# ── query(): Viết điều kiện dưới dạng chuỗi ─────────────────────────
#   Ưu điểm: Dễ đọc khi có nhiều điều kiện phức tạp

ket_qua = df_clean.query("Gia_Ban >= 1000000 and Is_VIP == True")
print(f"=== KHÁCH VIP MUA HÀNG >= 1 TRIỆU ({len(ket_qua)} dòng) ===")
display(ket_qua)

# ── Dùng biến bên ngoài trong query: dùng @tên_biến ─────────────────
nguong_gia = 500_000
ket_qua2 = df_clean.query("Gia_Ban > @nguong_gia and Danh_Muc == 'Gia dụng'")
print(f"\n=== GIA DỤNG GIÁ > {nguong_gia:,}đ ({len(ket_qua2)} dòng) ===")
display(ket_qua2)
```

## 5. Gom Nhóm & Thống Kê (Groupby + Aggregation)

`groupby` là một trong những công cụ mạnh nhất của Pandas. Nó hoạt động giống như
Pivot Table trong Excel: **Chia dữ liệu thành các nhóm → Áp dụng hàm thống kê → Gộp kết quả**.

### 5.1. Groupby cơ bản

```python
# ── Gom nhóm theo Danh_Muc, tính tổng So_Luong và Gia_Ban ───────────
print("=== TỔNG SỐ LƯỢNG VÀ GIÁ BÁN THEO DANH MỤC ===")
grouped = df_clean.groupby('Danh_Muc')[['So_Luong', 'Gia_Ban']].sum()
display(grouped)

# ── Tính nhiều chỉ số cùng lúc với agg() ────────────────────────────
print("\n=== THỐNG KÊ ĐA CHỈ SỐ THEO DANH MỤC ===")
thong_ke = df_clean.groupby('Danh_Muc').agg(
    Tong_So_Luong=('So_Luong', 'sum'),      # Tổng số lượng
    So_Luong_TB=('So_Luong', 'mean'),        # Trung bình số lượng
    Tong_Doanh_Thu=('Gia_Ban', 'sum'),        # Tổng giá bán (= doanh thu)
    Gia_TB=('Gia_Ban', 'mean'),               # Giá trung bình
    So_Don_Hang=('San_Pham', 'count'),         # Số đơn hàng (= số dòng)
    So_VIP=('Is_VIP', 'sum')                   # Số khách VIP (True=1, False=0)
).reset_index()  # reset_index() để đưa Danh_Muc từ index về cột bình thường

display(thong_ke)
```

### 5.2. Tạo cột mới — Doanh thu từng đơn hàng

```python
# ── Tạo cột Doanh_Thu = So_Luong * Gia_Ban ─────────────────────────
df_clean['Doanh_Thu'] = df_clean['So_Luong'] * df_clean['Gia_Ban']

print("=== 10 ĐƠN HÀNG CÓ DOANH THU CAO NHẤT ===")
display(df_clean.nlargest(10, 'Doanh_Thu')[['San_Pham', 'Danh_Muc', 'So_Luong', 'Gia_Ban', 'Doanh_Thu']])

# ── Groupby + Doanh thu ─────────────────────────────────────────────
print("\n=== TỔNG DOANH THU THEO DANH MỤC ===")
doanh_thu_theo_dm = df_clean.groupby('Danh_Muc')['Doanh_Thu'].sum().sort_values(ascending=False)
display(doanh_thu_theo_dm)
print(f"\n🏆 Danh mục dẫn đầu: {doanh_thu_theo_dm.idxmax()} ({doanh_thu_theo_dm.max():,.0f}đ)")
```

### 5.3. Groupby nâng cao — Gom nhóm theo nhiều cột

```python
# ── Gom nhóm theo Danh_Mục VÀ Is_VIP ───────────────────────────────
print("=== DOANH THU THEO DANH MỤC × LOẠI KHÁCH ===")
grouped_multi = df_clean.groupby(['Danh_Muc', 'Is_VIP']).agg(
    So_Don=('Doanh_Thu', 'count'),
    Tong_Doanh_Thu=('Doanh_Thu', 'sum'),
    Doanh_Thu_TB=('Doanh_Thu', 'mean')
).reset_index()
display(grouped_multi)

# ── Dùng pivot_table để tạo bảng chéo (cross-tab) ───────────────────
print("\n=== PIVOT TABLE: DOANH THU TRUNG BÌNH ===")
pivot = df_clean.pivot_table(
    values='Doanh_Thu',
    index='Danh_Muc',       # Dòng: Danh mục
    columns='Is_VIP',       # Cột: VIP / Không VIP
    aggfunc='mean'          # Hàm: trung bình
)
display(pivot)
```

```python
# ── Lời giải Challenge 4 ────────────────────────────────────────────

# Câu 1: Tổng doanh thu theo Danh Mục
print("=== CÂU 1: TỔNG DOANH THU THEO DANH MỤC ===")
cau1 = df_clean.groupby('Danh_Muc')['Doanh_Thu'].sum().sort_values(ascending=False)
print(cau1)

# Câu 2: Sản phẩm bán chạy nhất mỗi danh mục (theo số lượng)
print("\n=== CÂU 2: SẢN PHẨM BÁN CHẠY NHẤT MỖI DANH MỤC ===")
# Nhóm sản phẩm theo Danh_Muc + San_Pham, tính tổng So_Luong, rồi tìm max trong mỗi Danh_Muc
sp_grouped = df_clean.groupby(['Danh_Muc', 'San_Pham'])['So_Luong'].sum().reset_index()
idx_max = sp_grouped.groupby('Danh_Muc')['So_Luong'].idxmax()  # Vị trí max trong mỗi nhóm
cau2 = sp_grouped.loc[idx_max]
print(cau2.to_string(index=False))

# Câu 3: Tỉ lệ VIP trong từng danh mục
print("\n=== CÂU 3: TỈ LỆ KHÁCH VIP THEO DANH MỤC ===")
# Vì Is_VIP là True/False, mean() sẽ tính trung bình → tỉ lệ True
cau3 = df_clean.groupby('Danh_Muc')['Is_VIP'].mean().sort_values(ascending=False)
print((cau3 * 100).round(1).astype(str) + '%')
```

### 📖 Lời Giải Grand Challenge

```python
# ── Lời giải Grand Challenge ────────────────────────────────────────

# 1. Tạo cột Thang từ cột Ngay
df_clean['Ngay_dt'] = pd.to_datetime(df_clean['Ngay'])  # Chuyển sang datetime
df_clean['Thang'] = df_clean['Ngay_dt'].dt.month          # Lấy tháng (1-12)
print(f"1. Đã tạo cột Thang. Các tháng xuất hiện: {sorted(df_clean['Thang'].unique())}\n")

# 2. Doanh thu theo từng tháng
doanh_thu_thang = df_clean.groupby('Thang')['Doanh_Thu'].sum()
print(f"2. Doanh thu theo tháng:\n{doanh_thu_thang}\n")

# 3. Top 3 sản phẩm doanh thu cao nhất
top3_sp = df_clean.groupby('San_Pham')['Doanh_Thu'].sum().nlargest(3)
print(f"3. Top 3 sản phẩm doanh thu cao nhất:\n{top3_sp}\n")

# 4. Chi tiêu trung bình: VIP vs Không VIP
chi_tieu_tb = df_clean.groupby('Is_VIP')['Doanh_Thu'].mean()
print(f"4. Chi tiêu TB/đơn:\n  - Không VIP: {chi_tieu_tb.get(False, 0):,.0f}đ")
print(f"  - VIP:       {chi_tieu_tb.get(True, 0):,.0f}đ\n")

# 5. Danh mục có đơn hàng VIP nhiều nhất
vip_orders = df_clean[df_clean['Is_VIP'] == True]['Danh_Muc'].value_counts()
print(f"5. Số đơn VIP theo danh mục:\n{vip_orders}")
print(f"\n   Danh mục dẫn đầu: {vip_orders.idxmax()} ({vip_orders.max()} đơn)")
```
